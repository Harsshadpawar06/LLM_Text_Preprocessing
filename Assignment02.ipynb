PART 1 : Text Preprocessing

```python
# importing data from the Kaggle dataset
import pandas as pd
import re
data=pd.read_csv("testdata/twcs.csv")
```

```python
data.sample(5)
```

```python
# we will start with the lowercase conversion:
# 2.  lowercase / uppercase

data["text"]=data["text"].str.lower()
```

```python
# we will now remove html tags and urls:
import re
def remove_html_tags(text1):
    pattern=re.compile('<.*?>')
    return pattern.sub("",text1)
def remove_url(text):
    pattern=re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub("",text)
```

```python
data["text"]=data["text"].apply(remove_html_tags)
```

```python
data["text"]=data["text"].apply(remove_url)
```

```python
data.sample(5)["text"]
```

```python
# now we will be removing Emojis:
# 3.  Emojis

import emoji
def remove_emoji(text):
    clean_text=emoji.demojize(text)
    return clean_text
```

```python
data["text"]=data["text"].apply(remove_emoji)
```

```python
data.sample(5)["text"]
```

```python
# now we will remove the punctuation:
#4.  punctuation

import string
exclude=string.punctuation
def remove_punc(text):
    for char in exclude:
        text=text.replace(char,"")
    return text

def remove_punc1(text):
    return text.translate(str.maketrans("","",exclude))
```

```python
data["text"]=data["text"].apply(remove_punc)
```

```python
data["text"]=data["text"].apply(remove_punc1)
```

```python
data.sample(5)["text"]
```

```python
# we will update abbravation or slang words:
# 7.  abbravation

chat_words={
   " AFAIK":"As Far As I Know",
"AFK": "Away From Keyboard",
"ASAP":"As Soon As Possible",
"BTW":"By The Way",
"B4":"Before",
"LAMO":"Laugh My A.. Off",
"FYI":"For your information"
}
```

```python
def chat_conversion(text):
    
    new_text=[]
    words = text.split()
    for w in words:
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
#data["text"] = data["text"].apply(lambda x: ' '.join(x))
data["text"] = data["text"].apply(chat_conversion)
```

```python
data.sample(5)["text"]
```

```python
# removing stop words
# 6.  stopwords
import nltk
import re

nltk.download('stopwords')

def remove_stopwords(text, language='english'):
    # Load stopwords for the specified language
    stop_words = set(nltk.corpus.stopwords.words(language))

    words = text.split()
    cleaned_words = []

    for word in words:
        if word.lower() not in stop_words:
            cleaned_words.append(word)

    cleaned_text = ' '.join(cleaned_words)
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()  # Remove extra spaces

    return cleaned_text
```

```python
data["text"]=data["text"].apply(remove_stopwords)
```

```python
data.sample(5)["text"]
```

```python
# now we will tokaniize the words:
# 1.  Tokenization

import nltk

# Download the required NLTK resources
nltk.download('punkt')

# Define the function for tokenization
def tokenize_text(text):
    tokens = nltk.word_tokenize(text)
    return tokens
```

```python
data["text"]=data["text"].apply(tokenize_text)
```

```python
# At last we will doing steemming and lemmetization
# 8.  steemming and lemmetization

from textblob import TextBlob
from nltk.stem import PorterStemmer
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

# Function for stemming using NLTK's PorterStemmer
def stem_text(text):
    stemmer = PorterStemmer()
    stemmed_words = [stemmer.stem(word) for word in text.split()]
    return ' '.join(stemmed_words)

# Function for lemmatization using TextBlob
def lemmatize_text(text):
    # Join the list of words into a string before creating a TextBlob
    blob = TextBlob(' '.join(text))
    lemmatized_words = [word.lemmatize() for word in blob.words]
    return ' '.join(lemmatized_words)
```

```python
# Apply stemming to the "text" column and update it
#data["text"] = data["text"].apply(stem_text)
```

```python
# Apply lemmatization to the "text" column and update it
data["text"] = data["text"].swifter.apply(lemmatize_text)
```

Summarizing 5 advantages and disadvantages of One-Hot Encoding (OHE), Bag-of-Words (BOW), N-Grams, and TF-IDF:

### One-Hot Encoding (OHE)

**Advantages:**
1. Simple and intuitive representation.
2. Preserves information about distinct categories.
3. Works well with algorithms that expect numerical input.
4. No assumptions about relationships between categories.
5. Straightforward interpretation and implementation.

**Disadvantages:**
1. High dimensionality for large categorical spaces.
2. Not suitable for high cardinality categorical data.
3. Can result in a sparse matrix with many zeros.
4. May lead to multicollinearity issues in certain models.
5. Increases the complexity of the dataset.

### Bag-of-Words (BOW)

**Advantages:**
1. Simple and efficient representation of text data.
2. Works well for document classification tasks.
3. Scalable and easy to implement.
4. Can be used with various machine learning algorithms.
5. Good for tasks where word frequency matters.

**Disadvantages:**
1. Ignores word order and context.
2. Loses information about word relationships.
3. Treats all words as independent entities.
4. Doesn't capture semantics or word meanings.
5. Vocabulary size can be large and impact memory.

### N-Grams

**Advantages:**
1. Captures local word patterns and context.
2. Retains some sequential information.
3. Can improve performance for certain tasks.
4. Useful for tasks like machine translation.
5. Adjustability to task-specific contexts.

**Disadvantages:**
1. Increased dimensionality with higher N-grams.
2. Sensitivity to word order in training data.
3. Limited understanding of global context.
4. Sparse representation for higher N-grams.
5. Limited capturing of long-term dependencies.

### TF-IDF (Term Frequency-Inverse Document Frequency)

**Advantages:**
1. Considers word importance in a document collection.
2. Helps down-weight common words and emphasize rare ones.
3. Effective in information retrieval and search tasks.
4. Can be combined with various machine learning models.
5. Reduces the impact of stop words in the analysis.

**Disadvantages:**
1. Complexity increases with larger document collections.
2. May not work well for short documents.
3. Requires a corpus for meaningful IDF calculations.
4. Ignores word order and semantics.
5. Loss of some context information.
